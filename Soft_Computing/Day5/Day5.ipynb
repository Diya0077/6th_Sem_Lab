```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
elements = np.linspace(0, 10, 11) 
A = np.array([0.1, 0.3, 0.6, 0.8, 1.0, 0.9, 0.7, 0.5, 0.3, 0.2, 0.1])
B = np.array([0.2, 0.4, 0.5, 0.7, 0.9, 1.0, 0.8, 0.6, 0.4, 0.3, 0.2]) 
```

```python
def fuzzy_union(A, B):
 return np.maximum(A, B)
def fuzzy_intersection(A, B):
 return np.minimum(A, B)
def fuzzy_complement(A):
 return 1 - A
def fuzzy_difference(A, B):
 return np.maximum(A - B, 0) 
```

```python
union_result = fuzzy_union(A, B)
intersection_result = fuzzy_intersection(A, B)
complement_A = fuzzy_complement(A)
difference_result = fuzzy_difference(A, B)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(elements, A, 'bo-', label='Fuzzy Set A')
plt.plot(elements, B, 'go-', label='Fuzzy Set B')
plt.plot(elements, union_result, 'ro-', label='Union (A ∪ B)')
plt.plot(elements, intersection_result, 'yo-', label='Intersection (A ∩ B)')
plt.plot(elements, complement_A, 'co-', label='Complement of A (1 - A)')
plt.plot(elements, difference_result, 'mo-', label='Difference (A - B)')
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Values')
plt.title('Fuzzy Set Operations')
plt.legend(loc = 'best')
plt.grid(True)
plt.show()
```
