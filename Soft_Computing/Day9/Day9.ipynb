```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
logic_gates = {
'AND': {'X': np.array([[0, 0], [0, 1], [1, 0], [1, 1]]), 'y': np.array([0, 0, 0, 1])},
'OR': {'X': np.array([[0, 0], [0, 1], [1, 0], [1, 1]]), 'y': np.array([0, 1, 1, 1])},
'XOR': {'X': np.array([[0, 0], [0, 1], [1, 0], [1, 1]]), 'y': np.array([0, 1, 1, 0])}
}
```

```python

def perceptron(X, y, learning_rate=0.1, epochs=100):
     weights = np.random.rand(2)
     bias = np.random.rand()
     for epoch in range(epochs):
         for i in range(len(X)):
             output = np.dot(X[i], weights) + bias
             prediction = 1 if output >= 0 else 0
             error = y[i] - prediction
             weights += learning_rate * error * X[i]
             bias += learning_rate * error
     return weights, bias

# Train and display results
for gate, data in logic_gates.items():
     W, b = perceptron(data['X'], data['y'])
     print(f'Logic Gate: {gate}\nFinal Weights: {W}\nFinal Bias: {b}\n')
```
