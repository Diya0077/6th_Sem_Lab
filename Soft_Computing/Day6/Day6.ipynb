```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([0.1, 0.3, 0.6, 0.8, 1.0])
B = np.array([0.2, 0.5, 0.7, 0.9]) 
fuzzy_relation = np.minimum.outer(A, B) 
C = np.array([0.4, 0.6, 0.8, 1.0])
D = np.array([0.3, 0.5, 0.7])
second_relation = np.minimum.outer(C, D)
composition_result = np.maximum.reduce(np.minimum(fuzzy_relation[:, :, None],
second_relation[None, :, :]), axis=1) 
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
im1 = axes[0].imshow(fuzzy_relation, cmap='Blues', interpolation='nearest')
axes[0].set_title('Fuzzy Relation (A × B)')
fig.colorbar(im1, ax=axes[0])
im2 = axes[1].imshow(second_relation, cmap='Greens', interpolation='nearest')
axes[1].set_title('Fuzzy Relation (C × D)')
fig.colorbar(im2, ax=axes[1])
im3 = axes[2].imshow(composition_result, cmap='Reds', interpolation='nearest')
axes[2].set_title('Max-Min Composition Result')
fig.colorbar(im3, ax=axes[2])
plt.show() 
```
