```python
import numpy as np
```

```python
X = np.array([

[2.5, 2.4, 2.7],

[0.5, 0.7, 3.6],

[2.2, 2.9, 4.2],

[1.9, 2.2, 5.5],

[3.1, 3.0, 2.3]

])

# Step 2: Mean centering

X_meaned = X - np.mean(X, axis=0)

# Step 3: Covariance matrix

cov_matrix = np.cov(X_meaned, rowvar=False)

# Step 4: Eigen decomposition

eigen_values, eigen_vectors = np.linalg.eig(cov_matrix)

# Step 5: Sort eigenvalues and eigenvectors

sorted_index = np.argsort(eigen_values)[::-1]

sorted_eigenvalues = eigen_values[sorted_index]

sorted_eigenvectors = eigen_vectors[:, sorted_index]

# Step 6: Select top k eigenvectors (k=1 or 2)

k = 2

eigenvector_subset = sorted_eigenvectors[:, :k]

# Step 7: Transform the data

X_reduced = np.dot(X_meaned, eigenvector_subset)

# Output results

print("Covariance Matrix:\n", cov_matrix)

print("\nEigen Values:\n", eigen_values)

print("\nPrincipal Component (Top Eigenvector):\n", eigenvector_subset)

print("\nData after PCA (1D):\n", X_reduced)
```
